## Data Cleaning 

```python
matches.isnull().sum()
```

As we already seen earlier that column "city" have 7 null values,"winner" have 3 null values,"player_of_match" have 3 null values ,"umpire1" has 1 null value and "umpire2" has 1 null value.

Since since the data is important and removing of nan values cause loss of other column data so instead of dropping we replace the nan with suitable values and drop umpire3 and id column totally because there no value present in umpire3 column

```python
#since the umpire3 contains 100% null values so,we will drop the column umpire3
#other places where have only few null values replace by match_abandoned
matches.drop("umpire3",axis=1,inplace=True)
matches.drop("id",axis=1,inplace=True)
matches.city.fillna("match_abandoned",inplace=True)
matches.winner.fillna("no_result",inplace=True)
matches.player_of_match.fillna("no_result",inplace=True)
matches.dropna(inplace=True)
matches=matches.reset_index(drop=True)

matches.head()
```

```python

matches.isnull().sum()
```

```python

matches["team2"].value_counts()
```

We observe from above that same team appears twice with different names i.e. "Rising Pune Supergiants" and "Rising Pune Supergiant", therefore we need to fix this error

```python
## since we get few mistakes in team name of team which result two team names we combine both and
## give a particular name as Rising Pune Supergiants
matches["team2"]=matches["team2"].replace("Rising Pune Supergiant","Rising Pune Supergiants")
matches["team1"]=matches["team1"].replace("Rising Pune Supergiant","Rising Pune Supergiants")
matches["winner"]=matches["winner"].replace("Rising Pune Supergiant","Rising Pune Supergiants")
matches["toss_winner"]=matches["toss_winner"].replace("Rising Pune Supergiant","Rising Pune Supergiants")
```

As i found two names of same team in team1,team2,winner,and toss_winner column so I  replace  the Rising Pune Supergiant with Rising Pune Supergiants from all the columns

## Univariate Analysis

### No. of matches played in different season

```python
plt.figure(figsize=(16,8))
sns.countplot("season",data=matches,palette="bright")
plt.title("Number of Matches played in each IPL season",fontsize=15)
plt.xlabel("years",fontsize=15)
plt.ylabel("No. of matches",fontsize=15);
```

In year 2013 played maximum matches in a season in the history of ipl till 2017

```python
#top 3 teams who wins maximum matches till now in the history of ipl
df=sns.countplot(matches['winner'])

plt.xticks(rotation=90)
plt.show()
```

Mumbai Indian,Chennai SuperKings and Kolkata Knight Riders top 3 teams who won maximum number of matches 



```python
##visualise the difference between won of matches on the basis of batting first and bowling first
matches["batting_bowling_first"]=np.where(matches["win_by_runs"]>0,"bat_first","bowl_first")
```

```python
plt.figure(figsize = (18,10))
sns.countplot('season',hue='batting_bowling_first',data=matches,palette='hsv')
plt.title("Numbers of matches won by batting and bowling first ",fontsize=20)
plt.xlabel("Season",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

In year 2008,2009,2011,2012,2013,2014,2016,2017 the team bowling first team wins more matches as compared to batting first
In year 2010 and 2015 the team batting first team wins more matches as compared to bowling first

```python
# we will plot pie chart on Toss decision
Toss=matches.toss_decision.value_counts()
labels=np.array(Toss.index)
sizes = Toss.values
colors = ['lime', 'red']
plt.figure(figsize = (10,8))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%',startangle=90)
plt.title('Toss result',fontsize=20)
#plt.axis('equal',fontsize=10)
plt.show()
```

Most of teams choose field first after winning the toss.

Might be dew factor would be reason behind their decision


```python
#Most of the time player of match in cricket 
plt.figure(figsize = (18,10))
top_players = matches.player_of_match.value_counts()[:10]
fig, ax = plt.subplots()
ax.set_ylim([0,20])
ax.set_ylabel("Count")
ax.set_title("Top player of the match Winners")
top_players.plot.bar()
sns.barplot(x = top_players.index, y = top_players, orient='v', palette="hsv");
plt.show();
```

```python
##Let see in which stadium maximum matches are played
plt.figure(figsize=[18,6],facecolor="lightyellow")
ax=sns.countplot('venue',data=matches,palette="winter")
plt.title('Stadium',fontsize=15)
plt.xlabel("Stadium name",fontsize=15)
plt.ylabel("No of matches",fontsize=15)
plt.xticks(rotation=90)
for p in ax.patches:
    ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
plt.show()
```
